## companies

```python
companies=list()
page = 0
for page in range(100):
    page += 1
    company_url = 'https://www.linkedin.com/search/results/companies/?origin=DISCOVER_FROM_SEARCH_HOME&page='+str(page)
    browser.get(company_url)
    companys = [(i.text,i.get_attribute('href')) for i in browser.find_element_by_css_selector('#ember56 > div > ul').find_elements_by_css_selector('a')][1::2]
    companies.extend(companys)
```

```python
from urllib.parse import urlencode
```

```python
def get_Url(keywords):
    params = {
        'keywords': keywords,
        'origin':'GLOBAL_SEARCH_HEADER',
        'q': 'blended'
    }
    base_url = 'https://www.linkedin.com/voyager/api/typeahead/hitsV2?'
    url = base_url + urlencode(params)
    return url
```
