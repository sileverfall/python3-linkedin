## companies

```python
companies=list()
page = 0
for page in range(100):
    page += 1
    company_url = 'https://www.linkedin.com/search/results/companies/?origin=DISCOVER_FROM_SEARCH_HOME&page='+str(page)
    browser.get(company_url)
    companys = [(i.text,i.get_attribute('href')) for i in browser.find_element_by_css_selector('#ember57 > div > ul').find_elements_by_css_selector('a')][1::2]
    companies.extend(companys)
```

```python
len(companies)
```

```python
companies
```

```python
with open('./companies_tw.txt','w') as f :
    for item in companies:
        f.write('%s' % item[0]+','+item[1]+'\n')
```
