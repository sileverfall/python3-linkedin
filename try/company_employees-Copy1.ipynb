## Company page

```python
with open('./companies_tw.txt','r') as f :
    a = f.read()
```

```python
a = a.split('\n')
```

```python
def get_companyDic(companyList):
    coms=list()
    for _,com in enumerate(companyList):
            c = com.split(',')
            company = c[0]
            company_url = c[-1]
            try:
                browser.get(company_url)
                ta = browser.find_element_by_xpath("//*[contains(text(), '全部 ')]")
                out_com={
                    'company':company,
                    'company_url':company_url,
                    'employee_size':ta.text.split(' ')[1],
                    'employee_page':ta.find_element_by_xpath("./..").get_attribute('href')
                }
                coms.append(out_com)
            except Exception as e:
                print(company)
                print(company_url)
                print(e)
    return coms
```

```python
%time CompanyList = get_companyDic(a)
```

```python
with open('./companies_tw_href.txt','w') as f :
    for item in CompanyList:
        f.write('%s' % item)
```
