## 有哪些上市櫃公司

```python
url = 'http://www.tej.com.tw/webtej/doc/uid.htm'
res = requests.get(url,headers = headers)
```

```python
print(res.apparent_encoding)
```

```python
res.encoding = 'Big5'
```

```python
soup = bs(res.text,"lxml")
```

```python
soup
```

```python
stocks = [i.text.split()[1::2] for i in soup.select('table:nth-child(4) > tr > td') if '類股' not in ','.join(i.text.strip().split())][1:]
```

```python
stocks_com=list()
for com in stocks:
    stocks_com.extend(com)
```

```python
len(stocks_com)
```

```python
overCounters = [i.text.split()[1::2] for i in soup.select('table:nth-child(6) > tr > td') if '類股' not in ','.join(i.text.strip().split())][1:]
```

```python
overCounters_com = list()
for com in overCounters:
    overCounters_com.extend(com)
```

```python
len(overCounters_com)
```

```python
full_cash = [i.text.split()[1::2] for i in soup.select('table:nth-child(10) > tr > td') if '類股' not in ','.join(i.text.strip().split())][1:]
```

```python
full_com = list()
for com in full_cash:
    full_com.extend(com)
```

```python
len(full_com)
```

```python
emerging = [i.text.split()[1::2] for i in soup.select('table:nth-child(12) > tr > td') if '類股' not in ','.join(i.text.strip().split())][1:]
```

```python
emerging_com = list()
for com in emerging:
    emerging_com.extend(com)
```

```python
len(emerging_com)
```

```python
companies = stocks_com + overCounters_com +full_com +emerging_com
```

```python
len(companies)
```

```python
companies
```

```python
with open('./companiesTW.txt','w') as f :
    for item in companies:
        f.write('%s\n' % item)
```
